```python
class Node():
    def __init__(self, data):
        self.prev = None
        self.next = None
        self.data = data

class Queue(object):
    def __init__(self):
        self.num_elements = 0
        self.head = None
        self.tail = None
    def enqueue(self, data):
        new_node = Node(data)
        if self.num_elements == 0:
            self.head = new_node
            self.tail = new_node
        else:
            self.head.prev = new_node
            new_node.next = self.head
            self.head = new_node
        self.num_elements += 1
    def dequeue(self):
        if self.num_elements == 0:
            return None
        elif self.num_elements == 1:
            value = self.head.data
            self.head = None
            self.tail = None
            self.num_elements -= 1
            return value
        self.tail.prev.next = None
        value = self.tail.data
        self.tail = self.tail.prev
        self.num_elements -= 1
        return value
    def reorder(self, data):
        if self.head.data == data:
            return
        elif self.tail.data == data:
            new_node = self.tail
            new_node.prev.next = None
            self.tail = new_node.prev
            new_node.prev = None
            self.head.prev = new_node
            new_node.next = self.head
            self.head = new_node
            return
        current_node = self.head.next
        while True:
            if current_node.data == data:
                current_node.prev.next = current_node.next
                current_node.next.prev = current_node.prev
                current_node.prev = None
                current_node.next = self.head
                self.head.prev = current_node
                self.head = current_node
                break
            current_node = current_node.next
```

```python
class LRU_Cache(object):
    def __init__(self, size_limit):
        self.num_elements = 0
        self.size_limit = size_limit
        self.cache = dict()
        self.LRU = Queue()
    def get(self, data):
        if data in self.cache:
            if self.num_elements == 1:
                return self.cache[data]
            self.LRU.reorder(self.cache[data])
            return self.cache[data]
        return -1
    def set(self, key, value):
        if self.num_elements == self.size_limit:
            value = self.LRU.dequeue()
            del self.cache[value]
            self.num_elements -= 1
        self.LRU.enqueue(key)
        self.cache[key] = value
        self.num_elements += 1
```
